```python
## import needed package
import random
from IPython.display import clear_output

## Function list:
### Select symbo function
def Start_func():
    if input('Play 1: Do you want to be X or O? ').upper() == 'X':
        return ['X','O'] ##Tuple maybe better in here ('X','O')
    else:
        return ['O','X']

### Player sequencer function
def Player_Sequence():      
    if random.randint(1,2) == 1:
        return 1
    else:
        return 2
    
### Ready function
def Player_ready():
    return input('Are you ready to play? Yes/No ').lower()


### Print game map
def print_map(play_list,player_symbo):
    clear_output() ## This function come from display package, and only work in Jupytor
    for i in list(map(int,list(play_list[1]))):
        input_dic[i] = player_symbo[0]
    for j in list(map(int,list(play_list[2]))):
        input_dic[j] = player_symbo[1]
    print(f"{input_dic[9]}|{input_dic[8]}|{input_dic[7]}"
          +"\n"+"-"*5+"\n"+
          f"{input_dic[6]}|{input_dic[5]}|{input_dic[4]}"
          +"\n"+"-"*5+"\n"+
          f"{input_dic[3]}|{input_dic[2]}|{input_dic[1]}")    
    ## Apart from writing whole print in one print function, write print function in different lines also work, refer to solution
    
### player sequence  
def player_seq(first_player,sequence):
    if (len(play_list[1])+len(play_list[2])) == 0: ## first round
        return first_player
    else:
        return sequence % 2 + 1 ## not first round, switch player
     
        
### Player input
def player_input(play_list,sequence):
    temp = input(f'Play{sequence}, please choose your next position (1-9)?')
    while (temp in (play_list[1] + play_list[2])):
        temp = input(f'{temp} has alreay been selected, please select another number (1-9)')
    return temp


## function for sort player input
def sort_list(_input):
    return ''.join(sorted(list(_input))) # Don't need this function anymore, since we changed the checking method


### Check winner
def check_winner(play_list):
    win_list = ['123','456','789','147','258','369','159','357']
    temp = False
    for i in win_list:
        count = 0
        for j in i:
            if j in play_list:
                count += 1
            if count == 3:
                temp = True
                break
        if temp == True:
            temp = True
            break
    return temp


### replay
def replay():
    return input("Do you want to replay the game? Yes/No").lower() == 'yes'
    
```

```python
## Start
### Welcome
print('Welcome to tic tac toe!\n')
game_on = True
while game_on:
    ### clean up the input
    input_dic={9:' ',8:' ',7:' ',6:' ',5:' ',4:' ',3:' ',2:' ',1:' '}
    play_list = {1:'',2:''}
    sequence = 1                     ## Variable
    winner = False
    ### Select symbo and sequence
    Player_Symbo = Start_func()      ## Variable
    
    first_player = Player_Sequence() ## Variable
    print(f'Player{first_player} will go first.') 
    
    ### If player ready to play
    if Player_ready() == 'no':
        break #Jump out of loop

    ### print map
    print_map(play_list,Player_Symbo)
    
    ### game start, continue until have winner
    while not winner:
        sequence = player_seq(first_player,sequence) ##Update Variable
        
        play_list[sequence] += player_input(play_list,sequence) ##Update Variable
        
        print_map(play_list,Player_Symbo)
        
        if max(len(play_list[1]),len(play_list[2])) < 3:
            continue
        else:
            winner = check_winner(play_list[sequence])
            if winner == True:
                print(f'Player{sequence} has won!')
                break
            elif len(play_list[1]) + len(play_list[2]) == 9:
                print('No one win.')
                break
    
    if not replay():
        break #Jump out of loop
        
```
